```python
x = np.linspace(0,10,100)
#linspace crea una linea che colleghi tutti i due numeri (0 e 10) creando tanti puntini quanti ne dichiari cioè 100, con numpy andiamo a importare tutte le funzioni come quella esponenziale
y = np.exp(+x)
plt.plot(x,y);
#plt.show()
```

```python
x = np.linspace(0,10,100)
#linspace crea una linea che colleghi tutti quesi punti che collegano le cooordinate, con numpy la libreria da cui andiamo a importare tutte le funzioni come quella esponenziale
y = np.exp(-x)
plt.plot(x,y);
#plt.show()
```
